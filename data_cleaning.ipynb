```python
import pandas as pd 

from hipe4ml.tree_handler import TreeHandler 
from matplotlib import pyplot as plt 
```

```python
th = TreeHandler('PlainTree50K_DCM_trdrichrec_12agev.root', 'plain_tree')
```

Save original df length

```python
df = th.get_data_frame()
original_df_len = len(df)
```

```python
rename_dict = { c : c.replace("Complex_","") for c in df.columns}
rename_dict 
```

rename the columns to get rid of the annoying 'complex' prefix

```python
df = df.rename(columns=rename_dict)
df.columns
```

Only take into account positive particles

```python
df = df.loc[df['q'] == 1] 
original_positive_df_len = len(df)
df
```

```python
df = df[['mass2', 'p', 'pT','rapidity',
       't', 'v_tof', 'pid'
       ]]
print(len(df))
df.dropna()
df
```

```python
set(df['pid'])
```

```python
legal_pids = [-13.0, -11.0, 211.0, 321.0, 2212.0]
df = df.loc[df['pid'].isin(legal_pids)]
set(df['pid'])
```

```python
df['pid'] = df['pid'].replace({
2212 : 0,
321 : 1,
-13 : 2,
-11: 2,
211: 2
})
df.head()
```

```python
df.to_csv('particle_data.csv', index=False)
```

```python
current_len = len(df)
print(f'Original data frame length: {original_df_len}')
print(f'Original data frame length (positive particles only): {original_positive_df_len}')
print(f'Current data frame length: {current_len}')
print(f'Particles lost: {original_positive_df_len - current_len}')
print(f'percentage of particles kept: {100* current_len/original_positive_df_len:.2f}%')
```
