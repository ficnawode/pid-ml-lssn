```python
import torch
import torch.nn as nn
import pandas as pd 

from hipe4ml.tree_handler import TreeHandler 
from matplotlib import pyplot as plt 
import matplotlib
```

```python
df = pd.read_csv('particle_data.csv')
```

```python


fig, _ = plt.subplots(figsize=(8,6), dpi=200)

print(len(df))
plt.hist2d(df['p'], df['mass2'], bins=200, norm=matplotlib.colors.LogNorm(), range=[(0,12), (-1,3)])

x_axis_name = r"sign($q$) $\cdot p$ (GeV/c)"
y_axis_name = r"$m^2$ $(GeV/c^2)^2$"

plt.xlabel(x_axis_name, fontsize=9, loc="right")
plt.ylabel(y_axis_name, fontsize=9, loc="top")   

fig.tight_layout()
plt.show()
```
